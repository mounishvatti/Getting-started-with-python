```python
# #variable length parameters
def variable(*v):
    n=int(input("Enter the index: "))
    print("The variable is: " + v[n])
variable("C++","java","python")
 
```

```python
# #integers    
def int(*i):
   for j in range(0,len(i)):
        print(i[j])
int(10,20,30,40)
```

```python
#add and subtract
def num(x,y):
    add=(x+y)
    sub=(x-y)
    return add,sub
print(num(10,20))
```

```python
#employee name and salary set to default
def fun(empname='mad',salary=25000):
    print("Employee name: ",empname)
    print("Salary: ",salary)
fun('mounish',35000)
fun('mad')
fun(30000)
```

```python
#even or odd 
def evenorodd(num):
    if num%2==0: 
        print("Even")
    else:
        print("Odd")
a=int(input("Enter a number: "))
evenorodd(a)
```

```python
#Passing a List as an Argument
def my_function(food):
  for x in food:
    print(x)
fruits = ["apple", "banana", "cherry"]
my_function(fruits)
```
